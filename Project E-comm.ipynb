### EDA (Exploratory Data Analysis):

```python
# Find out relation with Reached On time (ROT) with relevant features
```

#### a.	Mode vs ROT

```python
# Ship mode is not reached on time to placed order 
sns.countplot(x=x['Reached.on.Time_Y.N'],hue=x['Mode_of_Shipment'])
plt.show()
```

#### b.	Cust rating vs ROT

```python
# median customer rating similar across ROT values
plt.figure(figsize=(12,8))
sns.boxplot(x=x['Reached.on.Time_Y.N'],y=x['Customer_rating'],data=x,color='r')
plt.title('customer rating wise ROT')
plt.show()
```

#### c.	Cost of Prod vs ROT

```python
sns.boxplot(x=x['Reached.on.Time_Y.N'],y=x['Cost_of_the_Product'],data=x)
plt.title('customer rating wise ROT')
plt.show()
```

#### d.	Impo vs ROT

```python
# low and medium prodouct importances category is highly not reached on time. and show thire count below
sns.countplot(x=x['Product_importance'],hue=x['Reached.on.Time_Y.N'])
plt.show()
```

#### e.	Any other major relations with ROT

```python
# highly low and medium prodouct importances  shipment mode is "ship" and thire count is show below.
sns.countplot(x=x['Product_importance'],hue=x['Mode_of_Shipment'])
plt.show()
```

```python
# 0 to 10 dollor discount offer this prodouct is deliver on time. high discount  offered prodouct is not deliver on time . 
plt.figure(figsize=(12,8))
sns.stripplot(x=x['Mode_of_Shipment'],y=x['Discount_offered'], data=x, hue=x['Reached.on.Time_Y.N'])
plt.show()
```

```python
# here 2000 to 4000 weight grams prodouct is not reached on time placed
plt.figure(figsize=(12,8))
sns.stripplot(x=x['Mode_of_Shipment'],y=x['Weight_in_gms'], data=x, hue=x['Reached.on.Time_Y.N'])
plt.show()
```

```python
# warehouse block wise count ROT here is 'f' block count is high not reached on time
sns.countplot(x=x['Warehouse_block'],hue=x['Reached.on.Time_Y.N'])
plt.show()
```

```python
# past prodouct puraches count on ROT
sns.countplot(x=x['Prior_purchases'],hue=x['Reached.on.Time_Y.N'])
plt.title("prior_purchases vs reached on time")
plt.show()
```
